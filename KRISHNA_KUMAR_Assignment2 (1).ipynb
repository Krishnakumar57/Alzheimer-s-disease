```python

import numpy as np
import pandas as pd
```

**TASK 1:DATA PREPROCESSING**



```python
df=pd.read_csv('/content/House Price India.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x = df.drop('Price',axis=1)
x.head()
```

```python
y = df['Price']
y
```

**Splitting the Dataset into the Training set and Test set**

```python
from sklearn.model_selection import train_test_split
     
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.4,random_state=10)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
     
```

**TASK 2:BUILDING THE ANN MODEL**

```python

reg = Sequential()
reg.add(Dense(4,activation='relu'))   #Input layer
reg.add(Dense(18,activation='relu'))  #Hidden Layer
reg.add(Dense(7,activation='relu'))   #Hidden Layer
reg.add(Dense(13,activation='relu'))  #Hidden Layer
#reg.add(Dense(20,activation='relu'))  #Hidden Layer
reg.add(Dense(1,activation='linear')) #Output Layer
```

```python
reg.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

 Training the ANN on the Training set

```python
reg.fit(xtrain,ytrain,batch_size=10,epochs=22,validation_data=(xtest,ytest))
```

```python
reg.predict([[190545,126487,894592,589520,958900,4]])
```
